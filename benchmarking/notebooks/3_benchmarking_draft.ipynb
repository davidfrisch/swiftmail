```python
import os
import json
from matplotlib import pyplot as plt
from pysimilar import compare

draft_benchmark_path = "../draft_emails/results/benchmark_results_responses_1722963476.4437811.json"
if not os.path.exists(draft_benchmark_path):
    raise FileNotFoundError("Please run the extraction benchmark first")
  
with open(draft_benchmark_path, "r") as file:
    data_answers = json.load(file)
```

```python
# length of the draft 

lengths = [len(data_answers["results"][draft]["generated_text"]) for draft in data_answers["results"]]

average_length = sum(lengths) / len(lengths)

plt.bar(range(len(lengths)), lengths)
plt.axhline(y=average_length, color='r', linestyle='--', label="Average Length")
plt.xlabel("Email Index")
plt.ylabel("Length of the Email (Characters)")
plt.xticks(range(len(lengths)), range(len(lengths)))
plt.title("Length of the Emails")
plt.show()
```

```python
time = [data_answers["results"][draft]["time"] for draft in data_answers["results"]]

average_time = sum(time) / len(time)

plt.bar(range(len(time)), time)
plt.axhline(y=average_time, color='r', linestyle='--', label="Average Time")
plt.xlabel("Email Index")
plt.ylabel("Time (Seconds)")
plt.xticks(range(len(time)), range(len(time)))
plt.title("Time taken to generate the Emails")
plt.show()
```
